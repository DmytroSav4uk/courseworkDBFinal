# [dbo].[AddLikeToPost]

```sql
/****** Object:  StoredProcedure [dbo].[AddLikeToPost]    Script Date: 12/22/2024 4:27:46 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[AddLikeToPost]
    @UserID INT,
    @PostID INT
AS
BEGIN
    -- Check if the user exists
    IF NOT EXISTS (SELECT 1 FROM Users WHERE UserID = @UserID)
    BEGIN
        PRINT 'User does not exist.';
        RETURN;
    END

    -- Check if the like already exists
    IF NOT EXISTS (
        SELECT 1
        FROM Likes
        WHERE UserID = @UserID AND PostID = @PostID
    )
    BEGIN
        -- Insert the like if it doesn't already exist
        INSERT INTO Likes (UserID, PostID, LikedAt)
        VALUES (@UserID, @PostID, GETDATE());

        PRINT 'Like added successfully.';
    END
    ELSE
    BEGIN
        PRINT 'User has already liked this post.';
    END
END;

GO
```

```sql
/****** Object:  StoredProcedure [dbo].[GetLikedPostsByUser]    Script Date: 12/22/2024 4:27:46 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[GetLikedPostsByUser]
    @UserID INT
AS
BEGIN
    SELECT 
        p.PostID, 
        p.UserID, 
        p.Content, 
        p.CreatedAt
    FROM Posts p
    INNER JOIN Likes l ON p.PostID = l.PostID
    WHERE l.UserID = @UserID
    ORDER BY l.LikedAt;
END;

GO
```

# [dbo].[GetMessagesBetweenUsers]

```sql
/****** Object:  StoredProcedure [dbo].[GetMessagesBetweenUsers]    Script Date: 12/22/2024 4:27:46 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[GetMessagesBetweenUsers]
    @UserID1 INT,
    @UserID2 INT
AS
BEGIN
    SELECT 
        m.MessageID, 
        m.SenderID, 
        m.ReceiverID, 
        m.Content, 
        m.SentAt
    FROM Messages m
    WHERE (m.SenderID = @UserID1 AND m.ReceiverID = @UserID2) 
       OR (m.SenderID = @UserID2 AND m.ReceiverID = @UserID1)
    ORDER BY m.SentAt;
END;

GO
```

# [dbo].[GetMostLikedPostPerMonth]

```sql
/****** Object:  StoredProcedure [dbo].[GetMostLikedPostPerMonth]    Script Date: 12/22/2024 4:27:46 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[GetMostLikedPostPerMonth]
    @SelectedYear INT
AS
BEGIN
    WITH MonthlyLikes AS (
        SELECT 
            PostID,
            MONTH(LikedAt) AS [Month],
            COUNT(*) AS [LikeCount]
        FROM Likes
        WHERE YEAR(LikedAt) = @SelectedYear
        GROUP BY PostID, MONTH(LikedAt)
    ),
    RankedPosts AS (
        SELECT 
            PostID,
            [Month],
            LikeCount,
            RANK() OVER (PARTITION BY [Month] ORDER BY LikeCount DESC) AS [Rank]
        FROM MonthlyLikes
    )
    SELECT 
        [Month],
        PostID,
        LikeCount
    FROM RankedPosts
    WHERE [Rank] = 1
    ORDER BY [Month];
END;

GO
```
